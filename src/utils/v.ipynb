```typescript
async function init () {
  const response = await fetch('https://clips.bliss.computer/f')
  const json = await response.json()
  console.log(json.map(f => f.timestamp))
  const visuals = json.reverse().map((f, i) => `Visual ${i+1}, ${new Date(f.timestamp).toLocaleDateString('en-US')} ${new Date(f.timestamp).toLocaleTimeString('en-US')}, https://clips.bliss.computer/data/${f.name}`)
  console.log(visuals.join('\n'))
}

init()
```
