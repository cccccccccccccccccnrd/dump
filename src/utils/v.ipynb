```typescript
async function init () {
  const response = await fetch('http://localhost:8782/f')
  const json = await response.json()
  const visuals = json.reverse().map((f, i) => `Visual ${i+1}, ${new Date(f.timestamp).toLocaleDateString('en-US')} ${new Date(f.timestamp).toLocaleTimeString('en-US')}, https://dump.cnrd.computer/data/${f.name}`)
  console.log(visuals.join('\n'))
}

init()
```
